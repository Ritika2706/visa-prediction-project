```python
import pandas as pd
df = pd.read_csv("G:/visa-prediction-project/outputs/final_model_ready_dataset.csv")
df.head()
```

```python
df["submission_date"] = pd.to_datetime(df["submission_date"])
df["decision_date"] = pd.to_datetime(df["decision_date"], errors='coerce')

df["submission_month"] = df["submission_date"].dt.month
df["submission_dayofweek"] = df["submission_date"].dt.dayofweek
df["submission_quarter"] = df["submission_date"].dt.quarter
```

```python
df["decision_binary"] = df["decision"].map({"Approved":1, "Rejected":0})
```

```python
categorical_cols = ["applicant_country", "visa_type", "processing_center", "season"]

df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
df.drop([
    "application_id",
    "submission_date",
    "decision_date",
    "processing_time_days",
    "processing_days_computed"
], axis=1, inplace=True, errors="ignore")
```

```python
print(df.info())
print(df.isnull().sum())
df.head()
```

```python
output_path = "../data/model_feature_engineered_dataset.csv"
df.to_csv(output_path, index=False)

print("ðŸŽ‰ Final dataset saved to:", output_path)
```
