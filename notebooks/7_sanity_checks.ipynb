```python
import pandas as pd

df = pd.read_csv("G:/visa-prediction-project/data/model_feature_engineered_dataset.csv")
print("\nColumns present in dataset:\n")
print(df.columns.tolist())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("\nSummary of numeric features:")
print(df[['processing_time_days_clean','application_year','submission_month','submission_dayofweek']].describe())
```

```python
print("\nDecision label balance:")
print(df['decision'].value_counts())
print("\nBinary label balance:")
print(df['decision_binary'].value_counts(normalize=True)*100)
```

```python
categorical_encoded_cols = [c for c in df.columns if "_" in c and (c.startswith("applicant_") or c.startswith("visa_type_") or c.startswith("processing_center_") or c.startswith("season_"))]
print("\nTotal encoded categorical features:", len(categorical_encoded_cols))
print("\nSample:", categorical_encoded_cols[:10])
```

```python
assert df['processing_time_days_clean'].min() >= 0, "Processing time has negative values!"
assert df['processing_time_days_clean'].max() < 1000, "Too large processing time — check calculation."
assert set(df['decision_binary'].unique()) <= {0,1}, "Decision encoding incorrect!"

print("\n✔ Basic sanity checks passed.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = df.corr(numeric_only=True)
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True)
plt.show()
```
